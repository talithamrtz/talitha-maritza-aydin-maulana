# Analisis Struktur File Saat Ini

Notebook ini akan membantu memecah `Portfolio_Talitha_Maritza.html` menjadi beberapa folder dan aset terpisah, termasuk `css/`, `assets/`, dan file HTML yang terorganisir.

Langkah:
- Identifikasi konten HTML, CSS, dan gambar bawaan.
- Tentukan struktur folder yang sesuai.
- Buat asset dan file baru.
- Perbarui referensi aset di `index.html`.
- Verifikasi struktur baru.

```python
from pathlib import Path

root = Path('e:/MINI PROJECT/AYANG PORTOFOLIO')
html_file = root / 'Portfolio_Talitha_Maritza.html'
print('HTML file exists:', html_file.exists())
text = html_file.read_text(encoding='utf-8', errors='ignore')
print('File size:', len(text), 'chars')
print('Contains <style>?', '<style>' in text)
print('Contains data:image?', 'data:image/' in text)
print('First 200 chars:\n', text[:200])
```

## Buat Folder Baru untuk Kode dan Aset

Sekarang buat folder yang jelas untuk struktur baru:
- `src/` untuk file HTML dan kode.
- `css/` untuk file stylesheet.
- `assets/images/` untuk gambar yang diekstrak dari `data:image/...`.
- `assets/` untuk aset statis lain bila diperlukan.


```python
dirs = [root / 'src', root / 'css', root / 'assets', root / 'assets' / 'images']
for d in dirs:
    d.mkdir(parents=True, exist_ok=True)
    print('created', d)
```

## Pindahkan File ke Folder yang Sesuai

Ekstrak `style` ke `css/styles.css`, pisahkan HTML ke `src/index.html`, dan simpan gambar ke `assets/images/` jika ada.


```python
import re
from base64 import b64decode

style_match = re.search(r'<style>(.*?)</style>', text, re.S)
if style_match:
    css_text = style_match.group(1).strip()
    (root / 'css' / 'styles.css').write_text(css_text, encoding='utf-8')
    print('Extracted CSS to css/styles.css')
else:
    print('No inline style block found')

# Extract embedded images
img_pattern = re.compile(r'src="(data:image/[^\"]+)"')
imgs = img_pattern.findall(text)
print('Found', len(imgs), 'embedded images')
new_text = text
for idx, img_data in enumerate(imgs, start=1):
    mime, b64data = img_data.split(',', 1)
    ext = mime.split('/')[1].split(';')[0]
    filename = f'image_{idx}.{ext}'
    path = root / 'assets' / 'images' / filename
    path.write_bytes(b64decode(b64data))
    print('Saved', filename)
    new_text = new_text.replace(f'src="{img_data}"', f'src="../assets/images/{filename}"', 1)

# Remove inline style from HTML and save to src/index.html
html_body = re.sub(r'<style>.*?</style>', '<link rel="stylesheet" href="../css/styles.css">', text, flags=re.S)
(src := root / 'src' / 'index.html').write_text(html_body, encoding='utf-8')
print('Saved new HTML to src/index.html')
```

## Perbarui Referensi dan Path

Pastikan referensi CSS dan gambar di `src/index.html` sudah benar:
- CSS: `../css/styles.css`
- Gambar: `../assets/images/image_*.png`

Juga periksa jika ada atribut `srcset` atau `manifest` lain yang perlu diperbarui.


```python
# Verify that resulting HTML references the new CSS path and images are updated
new_html = (root / 'src' / 'index.html').read_text(encoding='utf-8')
print('CSS link ok:', '../css/styles.css' in new_html)
if imgs:
    print('First image ref ok:', '../assets/images/image_1.' in new_html)
else:
    print('No images to verify')
```
